#### 1.Какие постулаты выдвинул Бор в теории атома водорода?
Постулат стационарных состояний. В атоме существуют стационарные состояния (не изменяющиеся со временем) находясь в которых атом не излучает энергии. Стационарным состояниям соответствуют определенные дискретные значения энергии и стационарные орбиты, по которым движутся
электроны. Движение электронов хотя и ускоренное, но электроны на этих орбитах не излучают энергию.
Правило частот. При переходе электрона с одного стационарного состояния в другое излучается (поглощается) один фотон с энергией ℎ𝜈 = 𝐸𝑛 − 𝐸𝑚.
𝐸𝑛, 𝐸𝑚- энергии стационарных состояний до излучения (поглощения) и после, соответственно. Если 𝐸𝑛 > 𝐸𝑚 - происходит испускание фотона – переход атома из состояния с большей энергией в состояние с меньшей энергией, т.е. переход электрона с более удаленной от ядра орбиты на более близлежащую. Если 𝐸𝑛 < 𝐸𝑚- происходит поглощение фотона – переход атома из
состояния с меньшей энергией в состояние с большей энергией, т.е. переход электрона на более удаленную от ядра орбиту. Набор различных дискретных частот 𝜈 = (𝐸𝑛−𝐸𝑚)/ℎ и определяет линейчатый спектр атома.

#### 2.Какие проблемы атомной физики и каким образом пытался  разрешить Н. Бор с помощью своих постулатов? 

Нильс Бор пытался разрешить противоречия между планетарной моделью атома Резерфорда и электродинамикой Максвелла. Дело в том, что, согласно законам электродинамики, заряды, движущиеся с ускорением, должны излучать электромагнитное излучение и при этом терять энергию. Электрон, двигающийся по орбите вокруг ядра, испытывает постоянное центростремительное ускорение, следовательно, он должен излучать электромагнитные волны, терять энергию, приближаться к ядру, и в конце концов за время порядка десятка микросекунд слиться с ним. Чтобы разрешить возникшие противоречия, Н. Бор предложил аксиомы-постулаты, объясняющие стабильность атомов и условия для излучения или поглощения энергии электронами в них.

#### 3.Как де Бройль объяснил правило Бора квантования орбит электрона в атоме водорода?
Де Бройль использовал представление о волнах де Бройля для наглядного толкования правила квантования Бора. Он рассмотрел волну де Бройля, бегущую вокруг ядра по круговой орбите электрона. Если на орбите длина волны 𝜆 укладывается
целое число раз, то волна при обходе вокруг ядра
будет всякий раз возвращаться в исходную точку
с той же фазой и амплитудой. В каждой точке орбиты установится неизменный колебательный
режим во времени и не возникнет излучения. В
этом случае орбита получится стационарной. Если указанное условие не выполняется, то при обходе вокруг ядра фаза и амплитуда волны не возвратятся к своим исходным данным и стационарного состояния не получится. Исходя из этого, де Бройль написал условие стационарности орбиты, или правило квантования, в виде:
$\frac{2\pi R}{\lambda}=n$

$\lambda = \frac{h}{p}$

$\frac{2\pi Rp}{h}=n, \ Rp = L \implies$ $L=n\hbar$

#### 4.Охарактеризуйте физическую сущность и значимость в атомной физике опытов Франка и Герца? 

Физическая сущность опытов Франка и Герца заключается в том, что они исследовали распределение электронов по скоростям после их соударения с атомами газа и убедились, что атомы способны воспринимать энергию электронов лишь порциями, равными энергиям перехода из основного в возбуждённое состояние.

Значимость опытов Франка и Герца в атомной физике заключается в том, что они дали прямое доказательство существования дискретных уровней энергии изолированных атомов, постулированных теорией Бора.

#### 5.Почему на вольтамперной характеристике лампы в опыте Франка-Герца наблюдаются неоднократные спады анодного тока?
Неоднократные спады анодного тока на вольтамперной характеристике лампы в опыте Франка-Герца наблюдаются из-за неупругих столкновений электронов с атомами газа. 
При таких столкновениях электроны почти полностью теряют свою энергию и не могут преодолеть задерживающего потенциала между сеткой и анодом. В результате электроны, потерявшие энергию, вылавливаются сеткой, и анодный ток резко падает.

#### 6.Почему анодный ток в опыте уменьшается в минимумах не до нуля? 

Анодный ток уменьшается в минимумах не до нуля, так как электроны после неупругих столкновений оседают на аноде, создавая небольшой остаточный ток.

#### 7.Какие столкновения электронов с атомами называются упругими, а какие – неупругими? На каких участках вольтамперной характеристики имеют место упругие, а на каких – неупругие столкновения?
Упругое столкновение - если сумма кинетических энергий двух частиц до соударения равна сумме кинетических энергий этих частиц после соударения, хотя и распределены эти энергии между ними по-другому.
Неупругое столкновение - если часть кинетической энергии пойдет на изменение внутреннего состояния одного из сталкивающихся тел

На вольтамперной характеристике упругие столкновения происходят до локального максимума I, а неупругие — после. Энергия электрона, налетающего на атом, недостаточна для того, чтобы перевести его в возбужденное состояние, и происходят только упругие столкновения. По мере увеличения разности потенциалов между сеткой и катодом энергия электронов увеличивается и оказывается достаточной для возбуждения атомов.

#### 8.С какой целью на коллектор подается задерживающее напряжение? Из каких соображений оно выбирается? 

Так как между коллектором и анодом поддерживается небольшое задерживающее напряжение (потенциал коллектора меньше потенциала анода), ток коллектора резко уменьшается, поскольку при неупругих соударениях с атомами электроны почти полностью теряют свою энергию и не могут преодолеть задерживающего потенциала. В этом случае электроны после неупругих столкновений оседают на аноде. Если энергия электронов после соударения достаточна для преодоления замедляющего потенциала в области ׀׀׀, то они попадут на коллектор C2. #### 10.Какие состояния атома называются стационарными? 

Стационарными называются не изменяющиеся со временем состояния атома, находясь в которых он не излучает энергии.

#### 11.Чем основное состояние атома отличается от возбужденного? 
Основное состояние атома — это состояние, в котором электрон(ы) находятся на самых низких энергетических уровнях, соответствующих минимальной энергии системы. В этом состоянии атом стабилен и не испускает энергию.

Возбужденное состояние атома возникает, когда один или несколько электронов переходят на более высокий энергетический уровень. В возбужденном состоянии атом имеет более высокую энергию и может быть менее стабильным. Электроны в возбужденном состоянии могут вернуться в основное состояние, при этом испуская энергию

#### 12.Что такое первый потенциал возбуждения? Какое состояние атома называется основным, а какое возбужденным? 

Первый потенциал возбуждения – разность потенциалов, сообщающая электрону энергию, равную энергии возбуждения атома.
Основное состояние – стационарное квантовое состояние с наименьшей внутренней энергией. 
Возбужденное состояние атома – это неустойчивое состояние атома с энергией, превышающей энергию основного (нулевого) состояния.

#### 13.Что такое потенциал ионизации? Чем он отличается от потенциала возбуждения? Какой из них больше, а какой меньше?
Потенциал ионизации — это разность потенциалов, которую должен пройти электрон, чтобы при столкновении ионизировать атом. Он равен энергии ионизации, которая равна работе вырывания электрона из атома или молекулы. 

Потенциал возбуждения — это разность потенциалов, сообщающая электрону энергию, равную энергии возбуждения атома. При передаче молекуле или атому энергии возможен не только полный отрыв электрона, но и переход электрона на более удаленную неустойчивую орбиту. 

Потенциал ионизации больше потенциала возбуждения.

## Предпроцессинг данных и подключение библиотек

Код подключает необходимые библиотеки и проводит обработку данных из таблиц с данными, полученными при эксперименте.

```python
from csv import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline


def file_read(path):
    with open(path, 'r') as file:
        data = file.readlines()
    data[0] = data[0][3:]
    #print(data)
    #data = data[1:]
    for i in range(len(data)):
        data[i] = data[i].replace(',','.')
        data[i] = data[i].split(';')
        for j in range(len(data[i])):
            if(data[i][j] != '' and data[i][j]!='\n'): 
                data[i][j] = float(data[i][j])
            else:
                data[i][j] = ''
    #print(data)
    return data
```

## Вычисление первого потенциала возбуждения атомов исследуемого газа

Для начала найдём значения напряжения на максимумах ВАХ.

Считывание данных из файла:

```python
table = file_read("изм49.1.csv")
table = np.array(table)
table = table.transpose()
table = table.tolist()
table = table[0]
print(table)
```

Вычисление $\Delta U$ и нахождение среднего значения

```python
dU = 0
for i in range(1,len(table)):
    dU += (table[i]-table[i-1])
dU /= len(table) - 1
print(dU)
```

По таблице это значение ближе всего к 11,6, таким образом, трубка заполнена аргоном.
